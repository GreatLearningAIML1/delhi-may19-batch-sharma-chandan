# 4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. 

```python
def Goal_diff_count(data):
    data['Goal_diff_count'] = data['GoalsFor'].astype(int) - data['GoalsAgainst'].astype(int)
    max_diff = data['Team'][(data['Goal_diff_count'] == data['Goal_diff_count'].max())]
    min_diff = data['Team'][(data['Goal_diff_count'] == data['Goal_diff_count'].min())]
    return data, max_diff, min_diff

    
data, max_diff, min_diff = Goal_diff_count(data)
print(max_diff)
print(min_diff)
print(data[['Team','GoalsFor','GoalsAgainst','Goal_diff_count']])
```
